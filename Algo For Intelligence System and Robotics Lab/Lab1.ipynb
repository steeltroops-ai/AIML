```python
def solve_8_queens(board, col):
    # base case: if all queens have been placed, return True
    if col >= len(board):
        return True

    for i in range(len(board)):
        # check if it's safe to place a queen at position (i, col)
        if is_safe(board, i, col):
            # place queen
            board[i][col] = 1

            # recurse to place the rest of the queens
            if solve_8_queens(board, col + 1):
                return True

            # backtrack
            board[i][col] = 0

    return False

def is_safe(board, row, col):
    # check row
    for i in range(len(board)):
        if board[row][i] == 1:
            return False

    # check diagonal
    for i, j in zip(range(row, -1, -1), range(col, -1, -1)):
        if board[i][j] == 1:
            return False
    for i, j in zip(range(row, len(board), 1), range(col, -1, -1)):
        if board[i][j] == 1:
            return False

    return True

board = [[0 for i in range(8)] for j in range(8)]
solve_8_queens(board, 0)
print(board)
```
